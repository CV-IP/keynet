```python
import vipy
from keynet.system import Keynet
```

```python
im = vipy.calibration.testimage().show()
```

```python
(sensor, model) = Keynet(inshape=(3,512,512), 
                         net=None,
                         global_geometric='hierarchical_permutation', 
                         global_photometric='identity', 
                         local_geometric='identity', 
                         local_photometric='identity',                         
                         blocksize=64,
                         blockshape=(2,2), 
                         permute_at_level=[0,1,2])
```

```python
sensor.fromimage(im).encrypt().asimage().mindim(512, interp='nearest').show()    
```
